## Formato de las columnas

```python
# crear funcion para estandarización del texto
def format_sentence_case(df):
    """
    Convierte todo el texto en un DataFrame a formato oración: 
    primera letra en mayúscula, resto en minúsculas.
    """
    for col in df.select_dtypes(include=['object']).columns: 
        df[col] = df[col].astype(str).str.capitalize()  
    return df

df = format_sentence_case(df)

df.head()
```

### Columnas categóricas

```python
# revisar columna de objetivo financiero
df['financial_goals'].value_counts()
```

```python
# función para quitar los guiones bajos
def remove_underscores(goal):
    return goal.replace('_', ' ')

df['financial_goals'] = df['financial_goals'].apply(remove_underscores)
```

```python
# diccionario para agregar acentos
accent_dict = {
    'Jubilacion': 'Jubilación',
    'Educacion': 'Educación'
}

# reemplazar los valores con acentos usando el diccionario
df['financial_goals'] = df['financial_goals'].apply(lambda x: accent_dict.get(x, x))
```

```python
# revisar columna de tolerancia al riesgo
df['risk_tolerance'].value_counts()
```

```python
# revisar columna de preferencia financiera
df['financial_preference'].value_counts()
```

```python
# diccionario para agregar acentos
accent_dict2 = {
    'Inversion': 'Inversión'
}

# reemplazar los valores con acentos usando el diccionario
df['financial_preference'] = df['financial_preference'].apply(lambda x: accent_dict2.get(x, x))
```

```python
# revisar columna de opciones financieras
df['financial_options'].value_counts()
```

```python
# diccionario para reemplazar mayúsculas
capital_dict = {
    'Etfs': 'ETFS',
    'Cdf': 'CDF',
    'Cdears': 'CDEARS'
}

# reemplazar los valores usando el diccionario
df['financial_options'] = df['financial_options'].apply(lambda x: capital_dict.get(x, x))
```

```python
# revisar columna de retorno de inversión
df['investment_period'].value_counts()
```

```python
# función para reemplazar valores
def replace_investment_period(period):
    if '1 mes' in period:
        return 3  # Reemplazar 1 mes por 3 meses
    elif '5 meses' in period:
        return 6  # Reemplazar 5 meses por 6 meses
    elif '1 año' in period:
        return 9  # Reemplazar 1 año por 9 meses
    elif '10 años' in period:
        return 12  # Reemplazar 10 años por 12 meses
    return period 

df['investment_period'] = df['investment_period'].apply(replace_investment_period)
```
